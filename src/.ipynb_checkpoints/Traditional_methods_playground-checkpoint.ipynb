In this file, `image` refers to the PIL image. `img` refers to the numpy array

```python
# Open an Image
def open_image(path, in_rgb=True):
    newImage = Image.open(path)
    if in_rgb:
        return newImage.convert("RGB")
    return newImage

# Show from source
def show_form_source(path):
    open_image

# Save Image
def save_image(image, path):
    image.save(path, 'png')

# Create a new image with the given size
def create_image(i, j):
    image = Image.new("RGB", (i, j), "white")
    return image

# Get the pixel from the given image
def get_pixel(image, i, j):
    # Inside image bounds?
    width, height = image.size
    if i > width or j > height:
        return None

    # Get Pixel
    pixel = image.getpixel((i, j))
    return pixel

# create numpy array from image
def make_img(image):
    return np.array(image)

# create image from numpy array
def make_image(img):
    return Image.fromarray(img)

# get most used color value
def get_topk_pixels(image, k=5):
#     If the maxcolors value is exceeded, the method stops counting and returns None. 
#     The default maxcolors value is 256. To make sure you get all colors in an image, 
#     you can pass in size[0]*size[1] 
#     (but make sure you have lots of memory before you do that on huge images).
#     [::-1] reverses order
    return sorted(image.getcolors(np.prod(image.size)), key=lambda x: x[0])[-k:][::-1]

# Separates colors. Returns list of images and pixel locations.
def separate_colors(image, k=None):
    if not k:
        k = len(image.getcolors(np.prod(image.size)))
        
    colors = get_topk_pixels(image, k)
    colors_dict = dict((val[1], Image.new('RGB', image.size, (255,255,255))) 
                        for val in colors)
    pixel_dict = dict((img, []) for img in colors_dict.keys())
        
    pix = image.load()
    for i in range(image.size[0]):
        for j in range(image.size[1]):
            if pix[i,j] in colors_dict:
                colors_dict[pix[i,j]].putpixel((i,j),(0,0,0))
                pixel_dict[pix[i,j]].append((i, j))

    return [(color, colors_dict[color], pixels) for color, pixels in pixel_dict.items()]

# Parse through the images and separate the graphs that are lines.
def separate_lines(image, k=None, verbose=True):
    img_and_pix = separate_colors(image, k)
    colors, images, pixels = zip(*img_and_pix)
    
#      Assumes:
#         line has #pixels >= 3/4* the min of width or height (to eliminate noise)
#         line has #pixels <= 1/4 the screen size (to eliminate background)
#         the line is moves at least 15 pixels in the y direction (or is 15 pixels thick)
#         len(pixels[i]) = how many non-white pixels are in image.

    trend_threshold = 100 # pixels
    variable_threshold = 200 # range between min and max
#
    inferred_lines = []
    labels = []
    for i, img in enumerate(images):
        img = crop_title(img)
        pix = pixels[i]
        x,y = zip(*np.array(pix))
        if len(pix) >= min(img.size) and len(pix) <= 0.25*np.prod(img.size) and len(set(y))>15:
            d = np.diff(pix, axis=0)
            segdists = np.sqrt((d ** 2).sum(axis=1))
            if(sum(segdists)/len(segdists))<5:
                if verbose:
                    display(img)
#                 img.save('tmp.png')
#                 text = pytesseract.image_to_string(Image.open('tmp.png'))
#                 print(text)
#                 os.remove('tmp.png')
                inferred_lines.append(img_and_pix[i])
                w, h = image.size
                actual_y = h - np.array(pix)[:,1]
                if actual_y[0] + trend_threshold < actual_y[-1]:
                    if verbose:
                        print('increasing')
                    labels.append('increasing')
                elif actual_y[0] - trend_threshold >  actual_y[-1]:
                    if verbose:
                        print("decreasing")
                    labels.append('decreasing')
                elif max(actual_y) - min(actual_y) > variable_threshold:
                    if verbose:
                        print('variable')
                    labels.append('variable')
                else:
                    if verbose:
                        print('stable')
                    labels.append('stable')
    
    return inferred_lines, labels

def crop_title(img):
    w, h = img.size
    top_line_and_words = get_line_and_word_boxes(img)[0]
    topright, bottomleft = top_line_and_words.position
    img.crop((0, bottomleft[1], w, h))
    return img
    
def get_line_and_word_boxes(img):
#   line.word_boxes is a list of word boxes (the individual words in the line)
#   line.content is the whole text of the line
#   line.position is the position of the whole line on the page (in pixels)

    line_and_word_boxes = tool.image_to_string(
        Image.open(filename), lang="eng",
        builder=pyocr.builders.LineBoxBuilder()
    )
    return line_and_word_boxes

def get_text(img):
    txt = tool.image_to_string(
        Image.open(filename),
        lang=lang,
        builder=pyocr.builders.TextBuilder()
    )
    return txt

def get_word_boxes(img):
#       box.content is the word in the box
#       box.position is its position on the page (in pixels)
    
    word_boxes = tool.image_to_string(
        Image.open(filename),
        lang="eng",
        builder=pyocr.builders.WordBoxBuilder()
    )
    return word_boxes


#     # digits is a python string
#     digits = tool.image_to_string(
#         Image.open(filename),
#         lang=lang,
#         builder=pyocr.tesseract.DigitBuilder()
#     )
```

```python
folder = '../data/'
labels = open(folder+'train_labels.txt', 'r').readlines()
graph_to_labels = {}
for r in labels:
    rlabels = r.rstrip().split(' ')
    graph_to_labels[rlabels[0]] = rlabels[1:]
```

```python
counter = 0
correct = 0
files = list(glob.glob(folder+"*.png"))
print("Predicting trends in a graph in no particular order...\n \n")
for file in files:
    counter += 1
    print("File number: ", counter)
    image = open_image(file)
    lines, trends = separate_lines(image, 10, False)
    predicted = Counter(trends).most_common()
    actual = Counter(graph_to_labels[file.split('/')[-1]]).most_common()
    print(file)
    print("Predicted: ", predicted)
    print("Actual:    ", actual)
    predicted.sort(key=lambda x: x[0])
    actual.sort(key=lambda x: x[0])
    print("Perfect?: ", predicted==actual)
    if predicted==actual:
        correct += 1
    print("\n")

print("Process finished.  Accuracy: {}/{}".format(correct, len(files)))
```
